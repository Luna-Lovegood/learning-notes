## 1. 图像的读取、显示、保存

参考：[python数字图像处理（2）：图像的读取、显示与保存](http://www.cnblogs.com/denny402/p/5121897.html)

```python
from skimage import io
import matplotlib.pyplot as plt

# 将显示的图像嵌入到 notebook 里
%matplotlib inline
```

可以通过 `skimage.io.imread(filename)` 函数来读取一幅图像，并通过 `skimage.io.imshow(image)` 函数来显示。

```python
# 读取一幅彩色图像并显示
img = io.imread("data/1/campus.jpg")  # numpy.ndarray
io.imshow(img)    # 实质上是通过 matplotlib 包绘制的
# plt.imshow(img) # 效果一样
plt.axis("off")   # 关闭坐标轴
io.show()
```

```python
# 读取一幅黑白图像
gray_img = io.imread("data/1/campus.jpg", as_grey=True)
io.imshow(gray_img)
plt.axis("off")   # 关闭坐标轴
io.show()
```

图像读入后是以 `numpy.ndarray` 的格式存储的，就是一个多维数组。可以用 `print` 打印一些图像的信息。

```python
print(type(img))  # 打印数据类型
print(img.shape)  # 打印图像的维度
print(img.size)   # 打印总像素个数
print(img.max())  # 打印最大像素值
print(img.min())  # 打印最小像素值
print(img.mean()) # 打印像素平均值
```

`skimage` 库自带了一些示例图片，如 `camera`、`rocket` 等，如果不想从外部读取图片，就可以使用这些示例图片。

```python
from skimage import data, data_dir
```

```python
print("The path of `data`: " + data_dir)
camera_img = data.camera()
# camera_img = io.imread(data_dir + "/camera.png")
io.imshow(camera_img)
plt.axis("off")   # 关闭坐标轴
io.show()
```

使用 `io` 模块的 `imsave(filename, arr)` 函数来保存图片。`filename` 参数表示保存的路径和名称，`arr` 参数表示需要保存的数组变量。

```python
save_img = data.chelsea()
io.imshow(save_img)
plt.axis("off")   # 关闭坐标轴
io.show()
io.imsave('data/1/save_chelsea.png', save_img) # 保存图片
```
